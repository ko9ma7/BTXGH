# Number of Posts Visualization
### 스압주의 - histogram

```python
def num_posts_agg(array, max=500):
    for file in array:
        try:
            plt.figure(figsize=(14,6))
            plt.hist(pd.read_csv(path+'/'+file, encoding='utf-8')['num_of_posts'], np.arange(0, 400, 20), color='coral', edgecolor="k")
            plt.xlabel("num_posts")
            plt.ylabel("num_followers")
            plt.xlim(0, max)
            plt.ylim(0, 1700)
            plt.title(file)
        except:
            pass
```

# Bio_Url & Private
### 스압주의 - simple bar charts

```python
ex1.head(1)
```

```python
def bio_private(array):
    for file in array:
        try:
            k = pd.read_csv(path+'/'+file, encoding='utf-8')
            
            public = (k['isprivate']==0).sum()*100/len(k)
            biourl = (1-(k['bio_url'].isna().sum())/len(k))*100
            rects = [public, biourl]
            
            
            plt.figure(figsize=(12,4))
            plt.bar(["Public Rate", "Bio_Url_Yes"], [public, biourl], color='plum', edgecolor="k")
            plt.xlabel("Followers")
            plt.ylabel("Ratio")
            plt.ylim(0, 100)
            plt.title(file)
            
            for i, v in enumerate(rects):
                plt.text(i-0.05, v+3, "{0:.2f}".format(v))
            
        except:
            pass
```
