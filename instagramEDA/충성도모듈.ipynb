## ID concentration
: 전체 코멘트 수 / 댓글러 유니크 아이디 개수

```python
from ast import literal_eval

commenterDf = pd.read_csv('data/influencer_commenters.csv') # 채린님이 합산한 인플루언서별 댓글러 리스트
commenterDf['cnt'] = commenterDf['commenters'].apply(lambda x: len(set(literal_eval(x))))
commenterDf.head()

df = pd.merge(df, commenterDf.loc[:, ['youtube','cnt']], on='youtube')
df['id_concentration'] = (df['avg_comment']*10)/df['cnt']
df.drop('cnt', inplace=True, axis=1)

df.to_csv('data/postNprofile_new.csv')
```
